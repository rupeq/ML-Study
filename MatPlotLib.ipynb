```python
x = np.linspace(-10, 10, 40)
y = x ** 2
x
```

```python
plt.plot(x, y);
```

```python
plt.plot(x, y, 'ro-'); # eq to color='red', marker='o', linestyle='--'
```

```python
plt.plot(x, y, color='red', marker='o', linestyle='--', label='First Line');
plt.plot(x, -y + 100, 'bx-', label='Second Line');
plt.grid();
plt.legend(loc='best');
```

```python
plt.plot(x, y, color='red', marker='o', linestyle='--', label='First Line');
plt.plot(x, -y + 100, 'bx-', label='Second Line');
plt.grid();
plt.legend(loc='upper center', ncol=2);

plt.xlabel('x', fontsize=14) # Ox label
plt.ylabel('y', fontsize=14) # Oy label

plt.ylim(0, 120) # Oy

plt.title(r'Graphic of functions $y = x^2 $ and $y = 100 - x^2 $', fontsize=16, y=1.05)
```

```python
t = np.linspace(0., 4 * np.pi, 101)
f = np.sin(t) + 3
plt.plot(t, f)

plt.axhline(3, color='lightgray', linestyle='--') # horisontal line on the middle
plt.text(10.5, 3.05, 'Middle') # cordinates of 'Middle'
plt.annotate('Local\nmin', xy=(3 * np.pi / 2, 2), # coordinates of 'Local min'
             xytext=(3.7, 2.7), # line has coordinates xy
             arrowprops=dict(arrowstyle='->', color='red')); # arrow
```

```python
# %load https://matplotlib.org/mpl_examples/pie_and_polar_charts/polar_bar_demo.py
"""
=======================
Pie chart on polar axis
=======================

Demo of bar plot on a polar axis.
"""
import numpy as np
import matplotlib.pyplot as plt


# Compute pie slices
N = 20
theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
radii = 10 * np.random.rand(N)
width = np.pi / 4 * np.random.rand(N)

ax = plt.subplot(111, projection='polar')
bars = ax.bar(theta, radii, width=width, bottom=0.0)

# Use custom colors and opacity
for r, bar in zip(radii, bars):
    bar.set_facecolor(plt.cm.viridis(r / 10.))
    bar.set_alpha(0.5)

plt.show()
```
