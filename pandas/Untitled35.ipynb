```python
df['total_kids']=df.children + df.babies
```

```python
df.groupby('hotel').agg({'total_kids':'mean'})
```

```python
df['has_kids']=df['total_kids']>0
```

```python
with_kids = df.query('is_canceled==1 and has_kids == True').shape[0]
```

```python
without_kids = df.query('is_canceled==1 and has_kids==False').shape[0]
```

```python
all_people = df.query('has_kids==False').shape[0]
```

```python
churn_rate_with_kids = with_kids / all_people * 100
```

```python
churn_rate_without_kids = without_kids / all_people * 100
```

```python
print(round(churn_rate_with_kids,2), round(churn_rate_without_kids, 2))
```
